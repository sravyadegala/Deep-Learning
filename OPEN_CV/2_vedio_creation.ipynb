```python
image = np.ones((700, 700, 3), np.uint8) * 224
# draw rectangle
img1=cv2.rectangle(image,(10,550),(120,590),(0,0,0),-1)
img1 = cv2.putText(image, "Start", (22,580), cv2.FONT_HERSHEY_SIMPLEX,0.75, (255,255, 255),2)
img1=cv2.rectangle(image,(70,510),(180,550),(0,0,0),-1) 
img1=cv2.rectangle(image,(130,470),(240,510),(0,0,0),-1)
img1 = cv2.putText(image, "Confidence", (132,499), cv2.FONT_HERSHEY_SIMPLEX,0.63, (255,255, 255),2)
img1=cv2.rectangle(image,(190,430),(300,470),(0,0,0),-1) 
img1=cv2.rectangle(image,(250,390),(360,430),(0,0,0),-1)
img1 = cv2.putText(image, "Patience", (260,420), cv2.FONT_HERSHEY_SIMPLEX,0.63, (255,255, 255),2)
img1=cv2.rectangle(image,(310,350),(420,390),(0,0,0),-1) 
img1=cv2.rectangle(image,(370,310),(480,350),(0,0,0),-1)
img1 = cv2.putText(image, "HardWork", (380,340), cv2.FONT_HERSHEY_SIMPLEX,0.63, (255,255, 255),2)
img1=cv2.rectangle(image,(430,270),(540,310),(0,0,0),-1) 
img1=cv2.rectangle(image,(490,230),(600,270),(0,0,0),-1)
img1 = cv2.putText(image, "Success", (500,255), cv2.FONT_HERSHEY_SIMPLEX,0.75, (255,255, 255),2)

'''======================================================================'''
# Goal
#1st layer
center = (620, 130)
axes = (75,100) 
angle = 0 
color = (0, 0, 255)  # Red
cv2.ellipse(img1, center, axes, angle, 0, 360, color, thickness=-1)

#2nd layer
center = (620, 130)
axes = (45,70)
angle = 0 
color = (255,255,255)   # white
cv2.ellipse(image, center, axes, angle, 0, 360, color, thickness=-1)

# 3rd layer
center = (620, 130)
axes = (25,40)  # Major and minor axes
angle = 0  # Rotation angle 
color = (0,0,255)   # Red
cv2.ellipse(image, center, axes, angle, 0, 360, color, thickness=-1)

# 4th layer
center = (620, 130)
axes = (10,20)
angle = 0 
color = (255,255,255)    # white
cv2.ellipse(image, center, axes, angle, 0, 360, color, thickness=-1)


values = []
def draw(event,x,y,flag,par):
    if event == cv2.EVENT_FLAG_LBUTTON:
        cv2.circle(img1,(x,y),3,(255,0,0),-1)
        values.append((x,y))
        if len(values)>=2:
            cv2.line(img1,values[-2],values[-1],(255,0,0),3)
            values.clear()
    if event == cv2.EVENT_FLAG_RBUTTON:
        cv2.circle(img1,(x,y),13,(255,0,0),3)

cv2.namedWindow('window',cv2.WINDOW_NORMAL)
cv2.setMouseCallback('window',draw)
count=1
while True:
    cv2.imshow('window', img1)
    key = cv2.waitKey(1) & 0xFF

    if key == ord('s'):
        filename = f"C:\\Users\\NAVYA\\Downloads\\MY_PYTHON_PRACTICE\\Module_3\\Deep_Learning\\env_dl\\Scripts\\Module_3\\Deep_Learning\\Open_CV\\images\\.png"
        cv2.imwrite(filename, img1)
        count += 1
    if key == ord('q'):
        break
cv2.destroyAllWindows()
# cv2.imwrite(r"C:\Users\sss\Pictures\cv_task\img1.png",img1)
```
